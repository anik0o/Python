```python
from google.colab import files
from google.colab import drive
import cv2
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import pyplot as plt

drive.mount('/gdrive')
dir_Path = '/gdrive/My Drive/'
img_name = 'kama.jpg'
```

```python
def test(dir_Path, img_name):
    file_Path = dir_Path + img_name
    img = cv2.imread(file_Path)
    rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    h, w, c = rgb_img.shape
    print(h, w, c) 
    plt.imshow(rgb_img)
    cv2.imwrite(dir_Path + 'kopia_' + img_name, rgb_img)
```

```python
def negatyw(dir_Path, img_name):
    file_Path = dir_Path + img_name
    img = cv2.imread(file_Path)
    rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    h, w, c = rgb_img.shape
    for x in range(0, w):
        for y in range(0, h):
            b, g, r = rgb_img[y, x]
            rgb_img[y, x] = 255-b, 255-g, 255-r
    cv2.imwrite(dir_Path + 'neg_' + img_name, rgb_img)
    plt.imshow(rgb_img)
```

```python
test(dir_Path, img_name)
```

```python
negatyw(dir_Path, img_name)
```

```python

def binary(dir_Path, img_name):
    file_Path = dir_Path + img_name
    img = cv2.imread(file_Path)
    im_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    ret, thresh = cv2.threshold(im_gray, 20, 255, cv2.THRESH_BINARY)
    contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    cv2.drawContours(img,contours,-1,(255,0,0),1)
    cv2.imwrite(dir_Path + 'contour_' + img_name, img)
    plt.imshow(img)

    perimeter = cv2.arcLength(contours[0], True)
    area = cv2.contourArea(contours[0])
    print("Obwód=",perimeter,"Powierzchnia=", area)
    
```

```python
binary(dir_Path, img_name)
```
